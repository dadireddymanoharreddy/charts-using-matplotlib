```python
# Import pandas and read the csv file
import pandas as pd
data = pd.read_csv("/content/cri t20.csv")
```

```python
# Display the first ten rows
data.head()
```

```python
data.describe()
```

```python
import matplotlib.pyplot as plt
x = data.Player
y = data.HS
plt.plot(x,y,'-s',color='b')
plt.title("Line Chart")
24
plt.ylabel("HS")
plt.xticks(rotation='vertical')
plt.xlabel("Player")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = data.Mat
plt.hist(x)
plt.title('Histogram')
plt.show()
```

```python
import matplotlib.pyplot as plt
x = data.Player
y = data.Inns
my_circle = plt.Circle( (0,0), 0.4, color='white')
plt.pie(y, labels=x, colors=['b','g','r','c','m','y','k','w','r','c','m','y','k','b','g','r','c','m','y'], wedgeprops={'linewidth':6,'edgecolor':'white'})       
plt.title("Cut Pie")
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```
